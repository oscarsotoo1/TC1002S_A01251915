```python
# Funcion que recibe un DataFrame, una proporcion y el nombre de la clase
# Genera cuatro conjuntos de datos para entrenamiento y pruebas del algoritmo de aprendizaje
from sklearn.model_selection import train_test_split

def split_label(dif, test_size, label):
    train, test = train_test_split(dif, test_size = test_size)
    features = dif.columns.drop(label)
    train_X = train[features]
    train_Y = train[label]
    test_X = test[features]
    test_Y = test[label]
    return train_X, train_Y, test_X, test_Y
```

```python
# Dividir el conjunto de datos de pasajeros del Titanic
train_X, train_Y, test_X, test_Y = split_label(dif, 0.2, "Price ($US)")
```

```python
# Aplicar onehotencoder a la columna "Embarked"
from sklearn.preprocessing import OneHotEncoder

one = OneHotEncoder(handle_unknown = 'ignore')

result = one.fit_transform(train_X["Lab"].values.reshape(-1, 1)).toarray()
train_X_1 = train_X
train_X_1[["0", "1", "2"]] = pd.DataFrame(result, index = train_X_1. index)
train_X_1
```

```python
# Aplicar escalamiento [0,1] a todos los valores
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
train_X_2 = min_max.fit_transform(train_X_1.values)
train_X_2 = pd.DataFrame(train_X_2, columns=train_X_1.columns)
train_X_2
```

```python


X = train_X_2
kmeans = KMeans(n_clusters=3).fit(X)
```
