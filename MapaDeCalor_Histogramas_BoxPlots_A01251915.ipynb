```python
# Importar librer√≠as a utilizar
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns
```

```python
# Carga del conjunto de datos de precios de casas de Boston
# de la librer√≠a sklearn

edu = pd.read_csv('Pace_of_life.csv', na_values = ':',
                   usecols = ["City","Walk","Bank","Talk","Heart"])
edu
```

```python
# Convertir el data set en DataFrame
import pandas as pd
df = pd.DataFrame(edu)
df.head()
```

```python
# Se agrega una columna al DataFrame con el dato objetivo
df['Walk'] = edu.target
df.head()
```

```python
# Verificar si hay datos perdidos
df.isnull().sum()
```

```python
# Se obtiene una matriz de correlaci√≥n
correlation_matrix = df.corr()
# Se usa la matriz de correlaci√≥n para dibujar un mapa de calor
# annot = True para mostrar los valores dentro del cuadro
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
datos= [edu["Walk"],edu["Bank"],edu["Talk"],edu["Heart"]]
fig = plt.figure(figsize=(10,7))

ax = fig.add_axes([0,0,1,1])

bp = ax.boxplot(datos)
plt.show()
```

```python
from collections import Counter
values = edu["Walk"]
histograma = Counter(min(valores//10*10,40) for valores in values)
plt.bar([x + 5 for x in histograma.keys()], 
        histograma.values(),
        10, 
        edgecolor = (0,0,0))
plt.xticks([10*i for i in range(11)])
plt.ylabel("Velocidad")
plt.show()     
```
